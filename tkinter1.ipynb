```python
from tkinter import*
root=Tk()
root.geometry('500x500')
root.iconbitmap('images.ico')
root.title('By Vipul')

scrollbar=Scrollbar(root)
scrollbar.pack(side=RIGHT ,fill=Y)
listbox=Listbox(root,yscrollcommand=scrollbar.set)
a=['vipul','yash','jeel','harpal']
for i in a:
    listbox.insert(END,i)
listbox.pack(fill="both",padx=22)
scrollbar.config(command=listbox.yview)


root.mainloop()
```

```python
from tkinter import*
root=Tk()
root.geometry('500x500')
root.minsize(500,500)
root.maxsize(500,500)
root.iconbitmap('images.ico')
root.title('By Vipul')


def update():
    print("Updating")
    root.geometry(f"{width.get()}x{height.get()}")
    root.minsize(width.get(),height.get())
    root.maxsize(width.get(),height.get())

Label(root, text="Window Resizer", font="comicsansms 15").pack()
print()
width=StringVar()
height=StringVar()

Entry(root,textvariable=width).pack()
Entry(root,textvariable=height).pack()

Button(root,text="apply",command=update).pack()


root.mainloop()
```
