```python
#5.7  Create a 3-D array with the shape of (2,3,4) and Keep integer numbers from 1 to 24.    Find the total EVEN numbers ADDITION from the given 3-D array. 
print('NAME:Jadav Vipul')
print('ENROLLMENT NO:21SOECE11031')
print("Class : 4CEA")

import numpy as np

# Create a 3D array with shape (2,3,4) and integers from 1 to 24
a = np.arange(1, 25).reshape(2, 3, 4)

# Find the even numbers in the array and add them up
even_sum = np.sum(a[a % 2 == 0])

# Print the array
print(a)

# Print the sum of even numbers
print("Sum of even numbers in the array: ", even_sum)
```

```python
#5.8 Create a 3-D array with the shape of (2,3,4) and Keep integer numbers from 1 to 24.  Find the total ODD numbers ADDITION from the given 3-D array.
print('NAME:Jadav Vipul')
print('ENROLLMENT NO:21SOECE11031')
print("Class : 4CEA")

import numpy as np

# Create a 3D array with shape (2,3,4) and integers from 1 to 24
a = np.arange(1, 25).reshape(2, 3, 4)

# Find the even numbers in the array and add them up
ODD_sum = np.sum(a[a % 2 != 0])

# Print the array
print(a)

# Print the sum of even numbers
print("Sum of odd numbers in the array: ", ODD_sum)
```
